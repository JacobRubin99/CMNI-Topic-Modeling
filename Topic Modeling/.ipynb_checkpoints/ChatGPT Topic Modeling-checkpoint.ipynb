```python
df = pd.read_csv('../Data/Questions_Final.csv')

# Comment out the following line if evaluating more than the CMNI
df = df[df["Scale"] == "Conformity to Masculine Norms Inventory"]

# Put reviews in a list
docs = df['Question Text'].to_list()

# Concatenate the documents into a single string
concatenated_docs = '\n'.join(docs)
```

```python
# Set ChatGPT API key.
# set openai.api_key variable here for local dev

# simple function that return just model response
# Limit the stochastic nature by setting temperature to 0 and top_p to 0
# Temperature is the level of randomness in the sampling that ChatGPT uses. If you limit the randomness, you get more deterministic.
# Top_p is the level of vocabulary that ChatGPT uses. If you limit the vocab, you get more deterministic.
def get_model_response(messages, 
                       model = 'gpt-4', 
                       temperature = 0, 
                       max_tokens = 2000,
                      top_p = 0):
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=temperature, 
        max_tokens=max_tokens, 
        top_p=top_p
    )

    return response.choices[0].message['content']
```

```python
from scipy.stats import chi2_contingency

# Assuming you have a dataframe named merged_df with "Original Factor" and "BERTopic Topic" columns
# Convert text values to categorical variables
merged_df["Factor"] = pd.Categorical(merged_df["Factor"])
merged_df["topic_name"] = pd.Categorical(merged_df["topic_name"])

# Create a contingency table
contingency_table = pd.crosstab(merged_df["Factor"], merged_df["topic_name"])

# Perform chi-squared test of independence to determine whether there is a stat. sig. assoc. 
# to see whether obs. freqs. in the contingency table differ significantly.
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print("Chi-squared statistic:", chi2)
print("P-value:", p_value)
print("Degrees of freedom:", dof)
print("Expected frequencies table:")
print(expected)
```
